```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2) # [1, 15]
```

```python
x = np.arange(1.0, 15.1, 0.1)
y = np.array([f(val) for val in x])
```

##### Многочлен первой степени, совпадающий с функцией f в точках 1, 15

```python
from scipy.linalg import solve
```

```python
coordinates_1 = np.array([1., 15.])
N = len(coordinates_1)

a = np.empty((N, 0))

for coordinate in coordinates_1:
    
    coefficients = []
    for power in range(N):
        coefficients.append(coordinate ** power)
        
    a = np.append(a, coefficients)
    
a = a.reshape((N, N))
b = np.array([f(x) for x in coordinates_1])
coef_1 = solve(a, b)

print(f"{a}\n")
print(f"{b}\n")
print(coef_1)
```

```python
def func_1(coef, x):
    return coef[0] + coef[1] * x
```

```python
plt.plot(x, y)
plt.plot(x, [func_1(coef_1, val) for val in x])
plt.show()
```

##### Многочлен второй степени, совпадающий с функцией f в точках 1, 8, 15

```python
coordinates_2 = np.array([1., 8., 15.])
N = len(coordinates_2)

a = np.empty((N, 0))

for coordinate in coordinates_2:
    
    coefficients = []
    for power in range(N):
        coefficients.append(coordinate ** power)
        
    a = np.append(a, coefficients)
    
a = a.reshape((N, N))
b = np.array([f(x) for x in coordinates_2])
coef_2 = solve(a, b)

print(f"{a}\n")
print(f"{b}\n")
print(coef_2)
```

```python
def func_2(coef, x):
    return coef[0] + coef[1] * x + coef[2] * (x ** 2)
```

```python
plt.plot(x, y)
plt.plot(x, [func_2(coef_2, val) for val in x])
plt.show()
```

##### Многочлен третьей степени, совпадающий с функцией f в точках 1, 4, 10, 15

```python
coordinates_3 = np.array([1, 4, 10, 15])
N = len(coordinates_3)

a = np.empty((N, 0)) #matrix of coefficients

for coordinate in coordinates_3: # each line of the matrix is an array of coefficients, 
                                 #which is power from 0 to N of each coordinate
    coefficients = []
    for power in range(N):
        coefficients.append(coordinate ** power)
        
    a = np.append(a, coefficients)
    
a = a.reshape((N, N))
b = np.array([f(x) for x in coordinates_3])
coef_3 = solve(a, b)

print(f"{a}\n")
print(f"{b}\n")
print(coef_3)
```

```python
def func_3(coef, x):
    return coef[0] + coef[1] * x + coef[2] * (x ** 2) + coef[3] * (x ** 3)
```

```python
plt.plot(x, y)
plt.plot(x, [func_3(coef_3, val) for val in x])
plt.show()
```
