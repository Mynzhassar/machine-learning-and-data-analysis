```python
path = "C:/ds/data_1_2/sentences.txt"
```

```python
file_obj = open(path, "r")
```

```python
data_list = list(file_obj)
data_list
```

```python
data_list_lower = []

for line in data_list:
    new_line = ""
    
    for char in line: new_line += char.lower()
    data_list_lower.append(new_line)
    
data_list_lower
```

```python
import re
data_splitted = []

for line in data_list_lower:
    line_splitted = re.split('[^a-z]', line)
    data_splitted.append(line_splitted)
    
print(data_splitted)
```

```python
data = []

for sentence in data_splitted:
    for word in sentence:
        
        if word != "": data.append(word)
            
print(data)
```

```python
dict_entrance = {}
pos = 0

for word in data:
    
    if word not in dict_entrance:
        dict_entrance[word] = pos
        pos += 1
```

```python
d, n = len(dict_entrance), len(data_splitted)
print(d, n)
```

```python
import pandas as pd
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.DataFrame(0, index=range(n),columns=range(d))
df.columns = dict_entrance.keys()

print(df.shape)
df
```

```python
for key in dict_entrance:
    
    for i, sentence in enumerate(data_splitted):

        for j, word in enumerate(sentence):
            if word == key: df.at[i, word] += 1
                
df
```

```python
from scipy.spatial.distance import cosine
```

```python
distance_from_first = []

for i in range(1, len(df)): distance_from_first.append(cosine(df.iloc[0], df.iloc[i]))
```

```python
ans_ind = []
```

```python
min_value = float("inf")
min_ind = -1

for i, val in enumerate(distance_from_first):
    
    if val < min_value:
        min_value = val
        min_ind = i
        
ans_ind.append(min_ind)
```

```python
min_value = float("inf")
min_ind = -1

for i, val in enumerate(distance_from_first):
    
    if val < min_value and i != ans_ind[0]:
        min_value = val
        min_ind = i
        
ans_ind.append(min_ind)
```

```python
print(ans_ind)
```

```python
with open("answer.txt", "w") as out:
    out.write(f"{ans_ind[0] + 1} {ans_ind[1] + 1}")
```
